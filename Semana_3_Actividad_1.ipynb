- Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
Los primeros almacenes de datos aparecieron a finales de la década de 1980 con el propósito de apoyar en el flujo de los datos, se tratan de repositorios centrales de información; es decir, una base de datos gigante que contiene gran diversidad de fuentes, y tiene la finalidad de tras ser procesada y analizada resolver problemáticas y tomar decisiones mejor informadas. Entre otros beneficios se encuentra el estar orientado al tema, integrado, no es volátil y mejorar el rendimiento.
Se constituye por las siguientes capas:
    - Nivel superior (Cliente frontend): Los resultados se exponen a través de informes, análisis y herramientas de minería de datos.
    - Nivel intermedio (Motor de análisis): Se adquieren el acceso a los datos y proceder al análisis
    - Nivel inferior (Servidor de la base de datos): Los datos se cargan y almacenan.
Existen dos tipos de almacenaje: 
    - Acceso frecuente: Almacenamiento muy rápido, como unidades SSD
    - Acceso poco frecuente: Almacenamiento de objetos rentable
Si se busca el diseñar un almacén de datos es preferible comenzar definiendo los requisitos comerciales, el alcance y esquematizar un diseño conceptual, acordes a las necesidades de los usuarios finales. Posteriormente, se procederá a crear un diseño lógico y físico para su almacenamiento, se establecerán procesos de transporte, respaldo y recuperación. Finalmente, se debe contemplar el espacio para la futura expansión y evolución.

**Solución 2: Rellenado de datos**

Copiamos el dataframe a otro para realizar la actividad

```python
dfrelleno = df.copy()
```

Se está calculando el promedio de la columna X12

```python
promedio = dfrelleno.X12.mean()
promedio
```

Se asignó el valor promedio calculado en todos los valores NaN de la columna X12 

```python
dfrelleno['X12'].fillna(value = promedio, 
                    inplace = True)
dfrelleno
```

Se asignó el valor de la mediana de la columna X13 en todos los valores NaN

```python
dfrelleno['X13'].fillna(value = dfrelleno.X13.median(), 
                    inplace = True)
dfrelleno
```

Sacamos la moda de la columna X2

```python
filter  = dfrelleno.X2.mode()
filter 
```

Se remplazaron los valores Nan por la moda en la columna X2

```python
dfrelleno['X2'].fillna(value = filter[0], #'NaN', 
                    inplace = True)
dfrelleno
```

Verificar si hay valores nulos
